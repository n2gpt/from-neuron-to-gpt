# 优化器类

上一章我们实现了自动微分：只需一行 `loss.backward()`，所有参数的梯度就自动计算完毕。但参数更新这一步，还是要我们手动写。随着网络层数增加，参数可能达到数百万，手动更新每一个显然不现实。

本章引入**优化器**（Optimizer）类，把参数更新这一步也自动化，完成整个框架最后一块拼图。

至此，框架三条链路的封装全部完成：

| 链路 | 负责的类 |
|------|----------|
| 前向传播 | **层**（Layer）|
| 梯度计算 | **张量**（Tensor）+ **损失函数**（Loss）|
| 参数更新 | **优化器**（Optimizer）|

```python
from abc import ABC, abstractmethod

import numpy as np
```

## 基础架构

### 张量

```python
class Tensor:

    def __init__(self, data):
        self.data = np.array(data)
        self.grad = np.zeros_like(self.data)
        self.gradient_fn = None
        self.parents = set()

    def backward(self):
        if self.gradient_fn is not None:
            self.gradient_fn()

        for parent in self.parents:
            parent.backward()

    @property
    def shape(self):
        return self.data.shape

    def __repr__(self):
        return f'Tensor({self.data})'
```

### 基础层

本章为基础层新增一个属性：**`parameters`**（参数列表）。

它返回本层所有需要参与梯度更新的张量。优化器通过这个接口，自动收集网络中所有层的参数，
而不需要知道每个层内部的具体结构。

```python
class Layer(ABC):

    def __call__(self, x: Tensor):
        return self.forward(x)

    @abstractmethod
    def forward(self, x: Tensor):
        pass

    @property
    def parameters(self):
        return []

    def __repr__(self):
        return f"{type(self).__name__}[]"
```

### 基础损失函数

```python
class Loss(ABC):

    def __call__(self, p: Tensor, y: Tensor):
        return self.loss(p, y)

    @abstractmethod
    def loss(self, p: Tensor, y: Tensor):
        pass
```

### 基础优化器

**基础优化器**（Base Optimizer）也是一个抽象类，定义了参数更新的统一接口。

创建一个优化器需要两样东西：

* **`parameters`**：所有需要更新的参数张量列表；
* **`lr`**（学习率）：每次更新时沿梯度方向移动的步长比例。

核心方法 `step()` 执行一次参数更新，具体的更新规则由子类实现。

优化器还有一个配套方法：`reset()`，在每轮训练开始前将所有参数的梯度清零。由于张量的 `grad` 在反向传播时累加，如果不清零，上一轮的梯度会叠加到下一轮，导致参数更新越来越大，偏离预期。这是深度学习训练中容易遗漏的关键步骤。

```python
class Optimizer(ABC):

    def __init__(self, parameters, lr):
        self.parameters = parameters
        self.lr = lr

    def reset(self):
        for p in self.parameters:
            p.grad = np.zeros_like(p.data)

    @abstractmethod
    def step(self):
        pass
```

## 数据

### 特征、标签

```python
feature = Tensor([28.1, 58.0])
label = Tensor([165])
```

## 模型

### 线性层

线性层覆盖了 `parameters` 属性，返回 `[self.weight, self.bias]`。
这样优化器就能自动找到需要更新的参数张量，不必知道层的内部结构。

```python
class Linear(Layer):

    def __init__(self, in_size, out_size):
        self.weight = Tensor(np.ones((out_size, in_size)) / in_size)
        self.bias = Tensor(np.zeros(out_size))

    def forward(self, x: Tensor):
        p = Tensor(x.data @ self.weight.data.T + self.bias.data)

        def gradient_fn():
            self.weight.grad += p.grad * x.data
            self.bias.grad += np.sum(p.grad, axis=0)
            x.grad += p.grad @ self.weight.data

        p.gradient_fn = gradient_fn
        p.parents = {x}
        return p

    @property
    def parameters(self):
        return [self.weight, self.bias]

    def __repr__(self):
        return f'Linear[weight{self.weight.data.shape}; bias{self.bias.data.shape}]'
```

### 均方误差损失函数

```python
class MSELoss(Loss):

    def loss(self, p: Tensor, y: Tensor):
        mse = Tensor(np.mean(np.square(y.data - p.data)))

        def gradient_fn():
            p.grad += -2 * (y.data - p.data)

        mse.gradient_fn = gradient_fn
        mse.parents = {p}
        return mse
```

### 随机梯度下降优化器

我们的第一个优化器类是**随机梯度下降优化器**（SGDOptimizer）。参数更新的规则非常简单：将每个参数减去学习率乘以对应的梯度。

$$
\theta \leftarrow \theta - \eta \cdot \frac{dL}{d\theta}
$$

其中 $\theta$ 代表任意参数（权重或偏置），$\eta$ 是学习率。

```python
class SGDOptimizer(Optimizer):

    def step(self):
        for param in self.parameters:
            param.data -= param.grad * self.lr
```

## 设置

### 超参数：学习率

```python
LEARNING_RATE = 0.00001
```

## 验证

### 建模

```python
layer = Linear(feature.shape[0], 1)
loss_fn = MSELoss()
optimizer = SGDOptimizer(layer.parameters, lr=LEARNING_RATE)
print(layer)
```

## 训练

### 完整训练步骤

有了优化器，一轮完整的训练现在固定为五步：

1. `reset()`：清零所有参数的梯度，防止上一轮梯度残留；
2. `forward()`：前向传播；
3. `loss_fn()`：计算损失值；
4. `backward()`：自动计算所有梯度；
5. `step()`：根据梯度更新所有参数。

这五步会在后续章节中作为训练循环的基本单元反复出现。

```python
# 1. 清零梯度
optimizer.reset()

# 2. 前向传播
prediction = layer(feature)
# 3. 计算损失值
loss = loss_fn(prediction, label)
print(f'loss: {loss}')

# 4. 自动微分（计算所有梯度）
loss.backward()

# 5. 更新参数
optimizer.step()
```

### 重新评估

```python
prediction = layer(feature)
loss = loss_fn(prediction, label)
print(f'loss: {loss}')
```

损失从 `14871.8` 降到了 `12503.0`，下降了约 `16%`。

至此，我们的神经网络训练框架的三条数据链路已经完整。
